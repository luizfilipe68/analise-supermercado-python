## 2. Importação de Bibliotecas

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 3. Carregamento dos Dados

```python
# Ajuste o caminho do arquivo
df = pd.read_csv('supermercado.csv')
df.head()
```

## 4. Entendimento Inicial dos Dados

```python
df.info()
df.describe()
df.isnull().sum()
```

## 5. Limpeza e Tratamento de Dados

```python
# remover duplicatas
df = df.drop_duplicates()

# exemplo de tratamento
# df = df.dropna()

df.head()
```

## 6. Estatística Descritiva

```python
media_preco = df['Preco_Normal'].mean()
desvio_preco = df['Preco_Normal'].std()
variancia_preco = df['Preco_Normal'].var()

print('Média:', media_preco)
print('Desvio padrão:', desvio_preco)
print('Variância:', variancia_preco)
```

## 7. Distribuição dos Dados

```python
plt.figure()
sns.histplot(df['Preco_Normal'])
plt.title('Distribuição de Preços')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Preco_Normal'])
plt.title('Boxplot de Preços')
plt.show()
```

## 8. Análise por Categoria

```python
categoria_qtd = df.groupby("Categoria")["title"].count().reset_index()
categoria_qtd
```

```python
import plotly.express as px

fig = px.bar(
    categoria_qtd,
    x="Categoria",
    y="title",
    title="Quantidade de produtos por categoria"
)

fig.show()
```

## 9. Relação entre Variáveis

```python
fig = px.scatter(
    df,
    x='Preco_Normal',
    y='Desconto',
    title='Preço vs Desconto'
)

fig.show()
```
